```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

datr = pd.read_csv('rocketHeights.csv')
table = datr.values

t=table[:,0]
h=table[:,1]

plt.figure(figsize = (10,8))
plt.plot(t, h, 'b.')

for i in range(1, 7):
    
    # get the polynomial coefficients
    y_est = np.polyfit(t,h, i)
    plt.subplot(2,3,i)
    plt.plot(t, h, 'o')
    # evaluate the values for a polynomial
    plt.plot(t, np.polyval(y_est, t))
    plt.title(f'Polynomial order {i}')

plt.tight_layout()
plt.show()
```
